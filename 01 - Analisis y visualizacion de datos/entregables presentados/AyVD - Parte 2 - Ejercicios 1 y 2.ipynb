## Ejercicio 1: Estimación

**Consigna:**  Calcular una estimación puntual y un intervalo de confianza de nivel (1-alpha) para la resta entre la media del salario Neto para Hombres menos la media del salario Neto para otros géneros(diferencia de las medias entre el grupoA y grupoB). 
¿Cómo se relaciona este intervalo de confianza con el test de hipótesis?

```
fig = plt.figure(figsize = (8,6))
seaborn.histplot(groupA)
seaborn.histplot(groupB, color = 'red')
```

Como la media es muy influenciada por datos atípicos, se procede a eliminar outliers.

```
groupA.to_frame()
groupB.to_frame()

k = 96.5
limite_max_A = groupA.quantile(k / 100)
limite_min_A = groupA.quantile((100-k)/100)
limite_max_B = groupB.quantile(k / 100)
limite_min_B = groupB.quantile((100-k)/100)

clean_groupA = groupA[(groupA < limite_max_A) & (groupA > limite_min_A)] 
clean_groupB = groupB[(groupB < limite_max_B) & (groupB > limite_min_B)] 
```

```
fig = plt.figure(figsize = (8,6))
seaborn.histplot(clean_groupA)
seaborn.histplot(clean_groupB, color = 'red')
```

```
mean_A = clean_groupA.mean()
mean_B = clean_groupB.mean()
std_A = clean_groupA.std()
std_B = clean_groupB.std()

nA = len(clean_groupA)
nB = len(clean_groupB)

print(f'La cantidad de hombres que cobran más de 100k es {nA}')
print(f'La cantidad de mujeres y personas de otros géneros que cobran más de 100k es {nB}')
```

El valor del estimador puntual de la diferencia de medias de salarios netos entre los grupos A y B es 16.342 pesos

```
print(f'El estimador puntual toma un valor de {mean_A - mean_B:.2f}')
```

```
# Estimación por intervalos
GL = round((((std_A**2)/nA + (std_B**2)/nB)**2) / ((((std_A**2)/nA)**2) / (nA - 1) + (((std_B**2)/nB)**2) / (nB - 1)))
print(f'Los grados de libertad son: {GL}')
```

```
t_025_GL = sts.t.ppf(1-alpha/2,GL) 
lim_inf =  (mean_A - mean_B) - t_025_GL* np.sqrt((std_A**2)/nA + (std_B**2)/nB)
lim_sup =  (mean_A - mean_B) + t_025_GL* np.sqrt((std_A**2)/nA + (std_B**2)/nB)
print(f'Los límites del intervalo son {lim_inf:.2f} y {lim_sup:.2f} pesos')
```

Se verifica que la estimación puntual de la diferencia de medias se encuentra dentro del intervalo de confianza

## Ejercicio 2: Test de hipótesis



### 2.1 Formalización

Describir formalmente los distintos componentes de un test de hipótesis para comprobar si la distribución de los salarios es distinta entre los grupos A y B.

**Hipótesis Nula**

H_0 =  No existe diferencia alguna entre las medias de los salarios de los grupos A y B

**Estadístico (Pivote)**
  * Identificar el estadístico: El estadístico es t
  * Escribir qué distribución tiene bajo $H_0$: Distribución t de Student


### 2.2 P-valor

1. Calcule el p-valor y decida si rechazar o no la hipótesis nula.
2. Interprete el resultado.

Links útiles:
* [Test de hipótesis usando scipy](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)
* [Test de Welch](http://daniellakens.blogspot.com/2015/01/always-use-welchs-t-test-instead-of.html)

```
sts.ttest_ind(clean_groupA, clean_groupB, equal_var = False)
```

### Conclusión
Un p-valor tan pequeño indica que la probabilidad de que la diferencia de medias de los salarios netos sea menor al valor estimado menos el estadistico es muy baja. Como la distribución t de Student es simétrica, lo mismo se aplica para la parte superior, es decir, la probabilidad de que la diferencia de medias sea mayor al valor estimado más el estadistico es muy baja. De esta manera, se rechaza la hipótesis nula por lo que la diferencia entre el valor de las medias salariales no es nula. En otras palabras, sí existe diferencia entre los salarios netos de los grupos A y B.  

### [Opcional] 2.3 Potencia del test

Nuestra muestra, ¿era lo suficientemente grande para detectar si existe o no una diferencia entre los grupos?

1. Utilice la función `tt_ind_solve_power` para calcular el tamaño necesario de la muestra para un poder estadístico de 0.8, 0.9 y 0.95, asumiendo una significancia estadística de 0.05.
2. ¿Cómo intepretan el poder estadístico de un test? Dado su conocimiento de dominio sobre los datos, ¿les parece que esta muestra es lo suficientemente grande para ser representativo de la tendencia general? ¿y para utilizarlo en un juicio penal contra una empresa XX por una causa de discriminación?

[Documentación](https://www.statsmodels.org/stable/generated/statsmodels.stats.power.tt_ind_solve_power.html)

NOTA: este análisis debería hacerse ANTES de recolectar los datos.

```
from statsmodels.stats.power import tt_ind_solve_power
```

```
effect_size = (groupA.mean() - groupB.mean()) / groupB.std()
# nobs1=None  - What we want to know
alpha = 0.05
ratio = len(groupB) / len(groupA)
```

```
# power = 0.8
power= 0.8
tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nA_teorico = tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nB_terico = ratio * nA_teorico
print(f'El tamaño de la muestra del grupo A debería ser {round(nA_teorico)}')
print(f'El tamaño de la muestra del grupo B debería ser {round(nB_terico)}')
```

```
# power = 0.9
power = 0.9
tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nA_teorico = tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nB_terico = ratio * nA_teorico
print(f'El tamaño de la muestra del grupo A debería ser {round(nA_teorico)}')
print(f'El tamaño de la muestra del grupo B debería ser {round(nB_terico)}')
```

```
# power = 0.95
power = 0.95
tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nA_teorico = tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=ratio)
nB_terico = ratio * nA_teorico
print(f'El tamaño de la muestra del grupo A debería ser {round(nA_teorico)}')
print(f'El tamaño de la muestra del grupo B debería ser {round(nB_terico)}')
```

El poder del test representa la probabilidad de que se rechace correctamente la hipótesis nula dado que es falsa. Como se puede ver, a mayor poder, mayor deben ser los tamaños de las muestras para que sean representativas de la tendencia general.
En el caso más expremo (poder = 0.95), la muestra del grupo A debe tener un tamaño de 493 y la del grupo B de 91. Como se vió en el ejercicio 1, los tamaños de las muestras superan ampliamente sus respectivos tamaños necesarios. Se concluye entonces que las muestras son lo suficientemente grandes como para ser representativas de la tendencia general.
